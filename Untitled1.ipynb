```python
import pandas as pd
import numpy as np
```

```python
movies_data = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits_data = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies_data.merge(credits_data,on = 'title')
```

```python
movies.shape
```

```python
movies.info()
```

```python
movies = movies[['genres','id','keywords','overview','title','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def helper(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
movies['genres'] = movies['genres'].apply(helper)
```

```python
movies['keywords'] = movies['keywords'].apply(helper)
```

```python
def cast(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
    
movies['cast']= movies['cast'].apply(cast)
```

```python
def crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
    
movies['crew']=movies['crew'].apply(crew)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[ i.replace(' ','') for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[ i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[ i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[ i.replace(' ','') for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['genres'] + movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
movies_df = movies[['id','title','tags']]
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:' '.join(x))
```

```python
movies_df
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

```python
movies_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
movies_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
similarity=cosine_similarity(vectors)
def recommend(movie):
    movie_index=movies_df[movies_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
    
    
recommend('Titanic')
```
